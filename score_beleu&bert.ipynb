```python
import pandas as pd
from nltk.translate.bleu_score import sentence_bleu
from bert_score import score
import sacrebleu

# Input Excel file
input_excel = "question.xlsx"  # Replace with your Excel file name

configurations = [
    {
        "sheet_name": "Eng",
        "Type":"Mysql",
        "query_column_translated": "Mresponse",
        "query_column_reference": "answer"
    },
    {
        "sheet_name": "Eng",
        "Type":"Neo4j",
        "query_column_translated": "Nresponse",
        "query_column_reference": "answer"
    },
]

# Output CSV
output_csv = "query_evaluation_eng.csv"

# Functions for BLEU, SacreBLEU, and BERTScore calculation
def calculate_bleu(reference, hypothesis):
    return sentence_bleu([reference.split()], hypothesis.split())

def calculate_sacrebleu(reference, hypothesis):
    bleu_score = sacrebleu.corpus_bleu([hypothesis], [[reference]])
    return bleu_score.score

def calculate_bertscore(reference, hypothesis):
    P, R, F1 = score([hypothesis], [reference], lang="en")
    return F1.mean().item()
```

```python
# Read data from multiple sheets, calculate scores, and save to CSV
def evaluate_queries(input_excel, configurations, output_csv):
    try:
        all_data = []  # List to store results from all sheets

        for config in configurations:
            sheet_name = config["sheet_name"]
            query_column_translated = config["query_column_translated"]
            query_column_reference = config["query_column_reference"]

            print(f"Processing sheet: {sheet_name} with columns '{query_column_translated}' and '{query_column_reference}'")

            try:
                # Read the Excel file for the specific sheet
                df = pd.read_excel(input_excel, sheet_name=sheet_name, header=0)

                # Process each row in the sheet
                for index, row in df.iterrows():
                    # Get translated and reference queries, handling NaN
                    translated_query = str(row.get(query_column_translated, "") or "")  # Replace NaN with an empty string
                    reference_query = str(row.get(query_column_reference, "") or "")  # Replace NaN with an empty string

                    # Compute metrics
                    try:
                        bleu = calculate_bleu(reference_query, translated_query)
                        sacre_bleu = calculate_sacrebleu(reference_query, translated_query)
                        bert_score = calculate_bertscore(reference_query, translated_query)
                    except Exception as e:
                        print(f"Error calculating scores for row {index} in sheet '{sheet_name}': {e}")
                        bleu, sacre_bleu, bert_score = None, None, None

                    # Append to all_data
                    all_data.append({
                        "Sheet Name": sheet_name,
                        "Row Index": index + 1,  # Add row index for traceability
                        "Translated Query": translated_query,
                        "Reference Query": reference_query,
                        "BLEU": bleu,
                        "SacreBLEU": sacre_bleu,
                        "BERTScore": bert_score
                    })

            except Exception as e:
                print(f"An error occurred while processing sheet '{sheet_name}': {e}")
                continue

        # Create a new DataFrame with all results
        results_df = pd.DataFrame(all_data)

        # Save the results to a CSV file
        results_df.to_csv(output_csv, index=False)
        print(f"All evaluation results saved to {output_csv}")

    except Exception as e:
        print(f"An error occurred: {e}")

# Run the evaluation
evaluate_queries(input_excel, configurations, output_csv)
```
